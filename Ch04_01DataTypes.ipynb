### 1. Declaring & Printing String

```python
S = "Spam"
print(S)
```

### 2. Length of the String

```python
print(len(S))
```

### 3. Accessing different alphabets in the string

```python
print(S[0]) # First element of string
print(S[-1]) # Last element of the string.
```

```python
#Slicing
S = "Spammer"
print(S[0:3]) # First to 3rd element
print(S[0:-1]) # First to 2nd last
```

### 4.String manpulation

Strings are immutable objects and their values cannot be directly changed after creating

```python
S = "Hello World"
#S[0] = "L" ## Does not work
S = "L" + S[1:]
print(S)
```

```python
S = "New World"
print(S + " Order")
```

```python
S = "Hello"
T = "World"
print(S+ "-"+ T)
```

### 5. String to list to string

```python
S = "Shrubbery"
L = list(S)
print(S)
```

```python
L[0] = "R"
print("".join(L))
```

```python
S = "This;is;a;new;world"
L =  S.split(';')
print(L) #A list of words is generated by Split method
K = " ".join(L)
print(K)
```

### 7. String Formatting

```python
print("%s and %s were best friends" %('Horse','Dog'))
```

```python
print('{1} was first man on moon, {0} was second man on moon'.format('Collins',"Armstrong"))
```

```python
print('{} was first man on moon, {} was second man on moon'.format('Collins',"Armstrong"))
```

Look at the above two statement and see how the order changed when the ordering was not given in the string
